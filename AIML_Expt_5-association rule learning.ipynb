```python
import pandas as pd
from collections import Counter
from itertools import combinations

def apriori_csv(filename, support_threshold):
    data = pd.read_csv(filename, header=None)

    # 2. Preprocess the transactions
    transactions = []
    for i in range(len(data)):
        transactions.append([str(item) for item in data.values[i] if str(item) != 'nan'])

    # 3. Initialize variables from the original code
    init = []
    for i in transactions:
        for q in i:
            if q not in init:
                init.append(q)
    init = sorted(init)

    s = int(support_threshold * len(init))

    # 4. Apriori algorithm implementation
    c = Counter()
    for i in init:
        for d in transactions:
            if i in d:
                c[i] += 1
    print("C1:")
    for i in c:
        print(str([i])+": "+str(c[i]))
    print()

    l = Counter()
    for i in c:
        if(c[i] >= s):
            l[frozenset([i])]+=c[i]
    print("L1:")
    for i in l:
        print(str(list(i))+": "+str(l[i]))
    print()
    pl = l
    pos = 1
    for count in range (2,1000):
        nc = set()
        temp = list(l)
        for i in range(0,len(temp)):
            for j in range(i+1,len(temp)):
                t = temp[i].union(temp[j])
                if(len(t) == count):
                    nc.add(temp[i].union(temp[j]))
        nc = list(nc)
        c = Counter()
        for i in nc:
            c[i] = 0
            for idx, transaction in enumerate(transactions):
                temp = set(transaction)
                if(i.issubset(temp)):
                    c[i]+=1
        print("C"+str(count)+":")
        for i in c:
            print(str(list(i))+": "+str(c[i]))
        print()
        l = Counter()
        for i in c:
            if(c[i] >= s):
                l[i]+=c[i]
        print("L"+str(count)+":")
        for i in l:
            print(str(list(i))+": "+str(l[i]))
        print()
        if(len(l) == 0):
            break
        pl = l
        pos = count
    print("Result: ")
    print("L"+str(pos)+":")
    for i in pl:
        print(str(list(i))+": "+str(pl[i]))
    print()
    
    # Association Rule Generation
    for l in pl:
        c = [frozenset(q) for q in combinations(l, len(l) - 1)]
        mmax = 0
        for a in c:
            b = l - a
            ab = l
            sab = 0
            sa = 0
            sb = 0
            for idx, transaction in enumerate(transactions):
                temp = set(transaction)
                if a.issubset(temp):
                    sa += 1
                if b.issubset(temp):
                    sb += 1
                if ab.issubset(temp):
                    sab += 1
            temp = sab / sa * 100 if sa != 0 else 0  # avoid division by zero
            if temp > mmax:
                mmax = temp
            temp = sab / sb * 100 if sb != 0 else 0  # avoid division by zero
            if temp > mmax:
                mmax = temp
            print(str(list(a)) + " -> " + str(list(b)) + " = " + str(sab / sa * 100) + "%")
            print(str(list(b)) + " -> " + str(list(a)) + " = " + str(sab / sb * 100) + "%")

        curr = 1
        print("choosing:", end=' ')
        for a in c:
            b = l - a
            ab = l
            sab = 0
            sa = 0
            sb = 0
            for idx, transaction in enumerate(transactions):
                temp = set(transaction)
                if a.issubset(temp):
                    sa += 1
                if b.issubset(temp):
                    sb += 1
                if ab.issubset(temp):
                    sab += 1
            temp = sab / sa * 100 if sa != 0 else 0  # avoid division by zero
            if temp == mmax:
                print(curr, end=' ')
            curr += 1
            temp = sab / sb * 100 if sb != 0 else 0  # avoid division by zero
            if temp == mmax:
                print(curr, end=' ')
            curr += 1
        print()
        print()
```

```python
# Example Usage:
filename = "Market_Basket_Optimisation.csv"  
support_threshold = 0.4  
apriori_csv(filename, support_threshold)
```
