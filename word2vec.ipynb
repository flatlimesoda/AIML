```python
import nltk
nltk.download("punkt")

from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize


sentences = [
    "Sustainable agriculture revolutionizes farming practices",
    "Crop rotation enhances soil fertility and yields",
    "Organic farming prioritizes environmental health and biodiversity",
    "Precision agriculture utilizes technology to optimize resource usage"
]

tokenized_sentences = [word_tokenize(sentence.lower()) for sentence in sentences]

model = Word2Vec(sentences = tokenized_sentences, vector_size = 100, window = 5, min_count = 1, workers = 4)

word_vector = model.wv['agriculture']

similar_words = model.wv.most_similar('agriculture', topn = 5)

print("vector representation of 'agriculture': ", word_vector)
print("Similar words to 'agriculture': ", similar_words)
```

```python
word_vectors = {word: model.wv[word] for word in model.wv.index_to_key}
similar_words_dict = {word: model.wv.most_similar(word, topn=3) for word in model.wv.index_to_key}

for word, vector in word_vectors.items():
    print(f"Vector for '{word}': {vector}")
for word, similar_words in similar_words_dict.items():
    print(f"Similar words to '{word}': {similar_words}")
```
