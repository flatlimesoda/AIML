```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
# triangle mf inbuilt function
temp = np.arange(0, 100, 1)
low_t = fuzz.trimf(temp, [10, 10, 20])
bavg_t = fuzz.trimf(temp, [15, 30, 45])
avg = fuzz.trimf(temp, [40, 50, 60])
aavg = fuzz.trimf(temp, [55, 70, 85])
high_t = fuzz.trimf(temp, [75, 90, 90])
fig, a =plt.subplots()
a.plot(temp, low_t, 'r', temp, bavg_t, 'y', temp, avg, 'g', temp, aavg, 'violet', temp, high_t, 'b')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('Temperature')
a.set_ylim(0, 1)
```

```python
pressure = np.arange(0, 6, 0.05)
low_p = fuzz.trimf(pressure, [1, 1, 1.75])
bavg = fuzz.trimf(pressure, [1.25, 2, 2.75])
avg = fuzz.trimf(pressure, [2, 3, 4])
aavg = fuzz.trimf(pressure, [3.25, 4, 4.75])
high_p = fuzz.trimf(pressure, [4.25, 5, 5.75])
fig, a =plt.subplots()
a.plot(pressure, low_p, 'r', pressure, bavg_p, 'y', pressure, avg, 'g', pressure, aavg, 'violet', pressure, high_p, 'b')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('Pressure')
a.set_ylim(0, 1)
```

```python
power = np.arange(0, 6, 0.05)
low = fuzz.trimf(power, [1, 1, 1.75])
mlow = fuzz.trimf(power, [1.25, 2, 2.75])
med = fuzz.trimf(power, [2.5, 3.25, 3.75])
mhigh = fuzz.trimf(power, [3.5, 4, 4.5])
high_h = fuzz.trimf(power, [4.25, 5, 5])
fig, a =plt.subplots()
a.plot(power, low, 'r', power, mlow, 'y', power, med, 'g', power, mhigh, 'violet', power, high_h, 'b')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('Power')
a.set_ylim(0, 1)
```

```python
f1 = np.array([22.5])
f2 = np.array([1.5])
r1 = fuzz.trimf(f1, [15, 30, 45])
r2 = fuzz.trimf(f2, [1.25, 2, 2.75])
r3 = fuzz.trimf(f1, [10, 10, 25])
r4 = fuzz.trimf(f2, [1, 1, 1.75])

mina = min(r1,r2)
minb = np.fmin(mina, mhigh)

minc = min(r3, r4)
mind = np.fmin(minc, high_h)

fig1, (a, b, c) = plt.subplots(nrows = 1, ncols = 3, figsize = (20,5), sharey = False)
a.plot(temp, bavg_t, 'b')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('Temperature')
a.set_ylim(0, 1.2)

b.plot(pressure, bavg, 'g')
b.set_ylabel('Fuzzy Membership')
b.set_xlabel('Pressure')
b.set_ylim(0, 1.2)

c.plot(power, minb, 'y')
c.set_ylabel('Fuzzy Membership')
c.set_xlabel('Power')
c.set_ylim(0, 1.5)
```

```python
fig1, (a, b, c) = plt.subplots(nrows = 1, ncols = 3, figsize = (20,5), sharey = False)
a.plot(temp, low_t, 'b')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('Temperature')
a.set_ylim(0, 1.2)

b.plot(pressure, low_p, 'g')
b.set_ylabel('Fuzzy Membership')
b.set_xlabel('Pressure')
b.set_ylim(0, 1.2)

c.plot(power, mind, 'y')
c.set_ylabel('Fuzzy Membership')
c.set_xlabel('Power')
c.set_ylim(0, 1.5)
```

```python
combined_heater_power = np.fmax(minb, mind)
fig2, (a, b, d) = plt.subplots(nrows = 1, ncols = 3, figsize = (20,5), sharey = False)

a.plot(power, minb, 'g')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('Power')
a.set_ylim(0, 1.5)

b.plot(power, mind, 'r')
b.set_ylabel('Fuzzy Membership')
b.set_xlabel('Power')
b.set_ylim(0, 1.5)

d.plot(power, combined_heater_power, 'b')
d.set_ylabel('Fuzzy Membership')
d.set_xlabel('Combined Heater Power')
d.set_ylim(0, 1.7)

plt.axvline(4.189013452914798, color = 'b', label = 'x*')
```
